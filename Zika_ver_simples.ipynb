# Load data

```python
def load_data(base_path,X_name,y_name):
#function to load the data split in .mat (Matlab saved matrix)
#the data used on this code was given in .mat matrix in the format
    mat_X=scipy.io.loadmat(base_path+X_name)
    mat_y=scipy.io.loadmat(base_path+y_name)
return(mat_X,mat_y)    
    
```

```python
C1_XI_4 , C1_yI_4 = load_data(cohort1_dir,'X_4d_inf','y_4d_inf')
C1_XI_7 , C1_yI_7 = load_data(cohort1_dir,'X_7d_inf','y_7d_inf')
C1_XN_4 , C1_yN_4 = load_data(cohort1_dir,'X_4d_ui','y_4d_ui')
C1_XN_7 , C1_yN_7 = load_data(cohort1_dir,'X_7d_ui','y_7d_ui')

C2_XI_4 , C2_yI_4 = load_data(cohort2_dir,'X_4d_inf','y_4d_inf')
C2_XI_7 , C2_yI_7 = load_data(cohort2_dir,'X_7d_inf','y_7d_inf')
C2_XI_10 , C2_yI_10 = load_data(cohort2_dir,'X_10d_inf','y_10d_inf')
C2_XN_4 , C2_yI_4 = load_data(cohort2_dir,'X_4d_ui','y_4d_ui')
C2_XN_7 , C2_yI_7 = load_data(cohort2_dir,'X_7d_ui','y_7d_ui')
C2_XN_10 , C2_yI_10 = load_data(cohort2_dir,'X_10d_ui','y_10d_ui')
```

```python
acuracia_infec_todosd=[]
acuracia_naoinfec_todosd=[]

infec_parcial_4d_todosd=[]
infec_parcial_7d_todosd=[]

naoinfec_parcial_4d_todosd=[]
naoinfec_parcial_7d_todosd=[]

acuracia_treino_infec_4d=[]
acuracia_treino_naoinfec_4d=[]
acuracia_treino_infec_7d=[]
acuracia_treino_naoinfec_7d=[]

usar_X_naoinfec_4d=C1_XN_4
usar_X_infec_4d=C1_XI_4
usar_Y_naoinfec_4d=C1_yN_4
usar_Y_infec_4d=C1_yI_4

usar_X_naoinfec_7d=C1_XN_7
usar_X_infec_7d=C1_XI_7
usar_Y_naoinfec_7d=C1_yN_7
usar_Y_infec_7d=C1_yI_7

usar_X_naoinfec_10d=C1_XN_10
usar_X_infec_10d=C1_XI_10
usar_Y_naoinfec_10d=C1_yN_10
usar_Y_infec_10d=C1_yI_10

numjanelas_lda=usar_X_infec_4d[0].size/lda_windows
if numjanelas_lda - np.floor(numjanelas_lda) >= 0.5:
    numjanelas_lda=np.ceil(numjanelas_lda)
else:
    numjanelas_lda=np.floor(numjanelas_lda)

numjanelas_lda=int(numjanelas_lda)

for rept in range(1,11):
    Xtreino=[]
    Ytreino=[]

    ###################    4D  ##########################
    tam_naoinfec= len(usar_X_naoinfec_4d)
    tam_infec=len(usar_X_infec_4d)
    ini_naoinfec=math.floor((rept-1)*tam_naoinfec/10)+1
    fim_naoinfec=math.floor(rept*tam_naoinfec/10)
    ini_infec=math.floor((rept-1)*tam_infec/10)+1
    fim_infec=math.floor(rept*tam_infec/10)
    if rept==10:
        fim_naoinfec=tam_naoinfec
        fim_infec=tam_infec

    Xval_naoinfec_4d=usar_X_naoinfec_4d[ini_naoinfec-1:fim_naoinfec,:]
    Xval_infec_4d=usar_X_infec_4d[ini_infec-1:fim_infec,:]
    Yval_naoinfec_4d=usar_Y_naoinfec_4d[ini_naoinfec-1:fim_naoinfec,:]
    Yval_infec_4d=usar_Y_infec_4d[ini_infec-1:fim_infec,:]

    Xtreino_naoinfec_aux1=usar_X_naoinfec_4d[0:ini_naoinfec,:]
    Xtreino_naoinfec_aux2=usar_X_naoinfec_4d[fim_naoinfec:,:]     
    Ytreino_naoinfec_aux1=usar_Y_naoinfec_4d[0:ini_naoinfec,:]
    Ytreino_naoinfec_aux2=usar_Y_naoinfec_4d[fim_naoinfec:,:]

    Xtreino_infec_aux1=usar_X_infec_4d[0:ini_infec,:]
    Xtreino_infec_aux2=usar_X_infec_4d[fim_infec:,:]  
    Ytreino_infec_aux1=usar_Y_infec_4d[0:ini_infec,:]
    Ytreino_infec_aux2=usar_Y_infec_4d[fim_infec:,:]



    Xtreino_4d=np.vstack((Xtreino_infec_aux1,Xtreino_naoinfec_aux1, 
                   Xtreino_infec_aux2,Xtreino_naoinfec_aux2,))

    Ytreino_4d=np.vstack((Ytreino_infec_aux1,Ytreino_naoinfec_aux1, 
                   Ytreino_infec_aux2,Ytreino_naoinfec_aux2,))
    
    ###################    7D  ##########################
    

    tam_naoinfec= len(usar_X_naoinfec_7d)
    tam_infec=len(usar_X_infec_7d)
    ini_naoinfec=math.floor((rept-1)*tam_naoinfec/10)+1
    fim_naoinfec=math.floor(rept*tam_naoinfec/10)
    ini_infec=math.floor((rept-1)*tam_infec/10)+1
    fim_infec=math.floor(rept*tam_infec/10)
    if rept==10:
        fim_naoinfec=tam_naoinfec
        fim_infec=tam_infec

    Xval_naoinfec_7d=usar_X_naoinfec_7d[ini_naoinfec-1:fim_naoinfec,:]
    Xval_infec_7d=usar_X_infec_7d[ini_infec-1:fim_infec,:]
    Yval_naoinfec_7d=usar_Y_naoinfec_7d[ini_naoinfec-1:fim_naoinfec,:]
    Yval_infec_7d=usar_Y_infec_7d[ini_infec-1:fim_infec,:]

    Xtreino_naoinfec_aux3=usar_X_naoinfec_7d[0:ini_naoinfec,:]
    Xtreino_naoinfec_aux4=usar_X_naoinfec_7d[fim_naoinfec:,:]     
    Ytreino_naoinfec_aux3=usar_Y_naoinfec_7d[0:ini_naoinfec,:]
    Ytreino_naoinfec_aux4=usar_Y_naoinfec_7d[fim_naoinfec:,:]

    Xtreino_infec_aux3=usar_X_infec_7d[0:ini_infec,:]
    Xtreino_infec_aux4=usar_X_infec_7d[fim_infec:,:]  
    Ytreino_infec_aux3=usar_Y_infec_7d[0:ini_infec,:]
    Ytreino_infec_aux4=usar_Y_infec_7d[fim_infec:,:]


    Xtreino_7d=np.vstack((Xtreino_infec_aux3,Xtreino_naoinfec_aux3, 
                   Xtreino_infec_aux4,Xtreino_naoinfec_aux4,))

    Ytreino_7d=np.vstack((Ytreino_infec_aux3,Ytreino_naoinfec_aux3, 
                   Ytreino_infec_aux4,Ytreino_naoinfec_aux4,))



    Xtreino_todosd=np.vstack((Xtreino_4d,Xtreino_7d))
    Ytreino_todosd=np.vstack((Ytreino_4d,Ytreino_7d))
    Xval_infec_todosd=np.vstack((Xval_infec_4d,Xval_infec_7d))
    Xval_naoinfec_todosd=np.vstack((Xval_naoinfec_4d,Xval_naoinfec_7d))
    Yval_infec_todosd=np.vstack((Yval_infec_4d,Yval_infec_7d))
    Yval_naoinfec_todosd=np.vstack((Yval_naoinfec_4d,Yval_naoinfec_7d))


    #preparacao para validacao do treino
    Xtreino_val_infec_4d=np.vstack((Xtreino_infec_aux1,Xtreino_infec_aux2))
    Ytreino_val_infec_4d=np.vstack((Ytreino_infec_aux1,Ytreino_infec_aux2))
    Xtreino_val_naoinfec_4d=np.vstack((Xtreino_naoinfec_aux1,Xtreino_naoinfec_aux2))
    Ytreino_val_naoinfec_4d=np.vstack((Ytreino_naoinfec_aux1,Ytreino_naoinfec_aux2))
    Xtreino_val_infec_7d=np.vstack((Xtreino_infec_aux3,Xtreino_infec_aux4))
    Ytreino_val_infec_7d=np.vstack((Ytreino_infec_aux3,Ytreino_infec_aux4))
    Xtreino_val_naoinfec_7d=np.vstack((Xtreino_naoinfec_aux3,Xtreino_naoinfec_aux4))
    Ytreino_val_naoinfec_7d=np.vstack((Ytreino_naoinfec_aux3,Ytreino_naoinfec_aux4))

    Ytreino_val_infec_4d=np.ravel(Ytreino_val_infec_4d)
    Ytreino_val_naoinfec_4d=np.ravel(Ytreino_val_naoinfec_4d)
    Ytreino_val_infec_7d=np.ravel(Ytreino_val_infec_7d)
    Ytreino_val_naoinfec_7d=np.ravel(Ytreino_val_naoinfec_7d)

    #fim da preparacao para validacao do treino

    Ytreino_todosd=np.ravel(Ytreino_todosd)
    Yval_naoinfec_4d=np.ravel(Yval_naoinfec_4d)
    Yval_infec_4d=np.ravel(Yval_infec_4d)
    Yval_naoinfec_7d=np.ravel(Yval_naoinfec_7d)
    Yval_infec_7d=np.ravel(Yval_infec_7d)
    Yval_naoinfec_todosd=np.ravel(Yval_naoinfec_todosd)
    Yval_infec_todosd=np.ravel(Yval_infec_todosd)

    Xtreino_4d_lda=[]
    Xtreino_7d_lda=[]
    Xval_infec_4d_lda=[]
    Xval_naoinfec_4d_lda=[]
    Xval_infec_7d_lda=[]
    Xval_naoinfec_7d_lda=[]
    Xval_infec_todosd_lda=[]
    Xval_naoinfec_todosd_lda=[]
    Xval_infec_4d_rf3_lda=[]
    Xval_naoinfec_4d_rf3_lda=[]
    Xval_infec_7d_rf3_lda=[]
    Xval_naoinfec_7d_rf3_lda=[]                
    Xtreino_val_infec_4d_lda=[]
    Xtreino_val_naoinfec_4d_lda=[]
    Xtreino_val_infec_7d_lda=[]
    Xtreino_val_naoinfec_7d_lda=[]

    
    for mult_lda in range(1,numjanelas_lda+1):
        comp_inicial_lda=num_fazer_lda*(mult_lda-1)

        if mult_lda==numjanelas_lda:
            comp_final_lda=usar_X_infec_4d[0].size
        else:
            comp_final_lda=num_fazer_lda*(mult_lda)

        tempLDA_Xtreino_todosd=Xtreino_todosd[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xval_infec_4d=Xval_infec_4d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xval_naoinfec_4d=Xval_naoinfec_4d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xval_infec_7d=Xval_infec_7d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xval_naoinfec_7d=Xval_naoinfec_7d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xval_infec_todosd=Xval_infec_todosd[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xval_naoinfec_todosd=Xval_naoinfec_todosd[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xtreino_val_infec_4d=Xtreino_val_infec_4d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xtreino_val_naoinfec_4d=Xtreino_val_naoinfec_4d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xtreino_val_infec_7d=Xtreino_val_infec_7d[:,comp_inicial_lda:comp_final_lda]
        tempLDA_Xtreino_val_naoinfec_7d=Xtreino_val_naoinfec_7d[:,comp_inicial_lda:comp_final_lda]

        lda_todosd=LDA(n_components=1)
        tempLDA_Xtreino_todosd_lda = lda_todosd.fit_transform(tempLDA_Xtreino_todosd, Ytreino_todosd)

        tempLDA_Xval_infec_todosd_lda=lda_todosd.transform(tempLDA_Xval_infec_todosd)
        tempLDA_Xval_naoinfec_todosd_lda=lda_todosd.transform(tempLDA_Xval_naoinfec_todosd)

        tempLDA_Xval_infec_4d_rf3_lda=lda_todosd.transform(tempLDA_Xval_infec_4d)
        tempLDA_Xval_naoinfec_4d_rf3_lda=lda_todosd.transform(tempLDA_Xval_naoinfec_4d)
        tempLDA_Xval_infec_7d_rf3_lda=lda_todosd.transform(tempLDA_Xval_infec_7d)
        tempLDA_Xval_naoinfec_7d_rf3_lda=lda_todosd.transform(tempLDA_Xval_naoinfec_7d)
        tempLDA_Xtreino_val_infec_4d_lda=lda_todosd.transform(tempLDA_Xtreino_val_infec_4d)
        tempLDA_Xtreino_val_naoinfec_4d_lda=lda_todosd.transform(tempLDA_Xtreino_val_naoinfec_4d)
        tempLDA_Xtreino_val_infec_7d_lda=lda_todosd.transform(tempLDA_Xtreino_val_infec_7d)
        tempLDA_Xtreino_val_naoinfec_7d_lda=lda_todosd.transform(tempLDA_Xtreino_val_naoinfec_7d)


        if mult_lda==1:
            Xtreino_todosd_lda=tempLDA_Xtreino_todosd_lda
            Xval_infec_todosd_lda=tempLDA_Xval_infec_todosd_lda
            Xval_naoinfec_todosd_lda=tempLDA_Xval_naoinfec_todosd_lda          
            Xval_infec_4d_rf3_lda=tempLDA_Xval_infec_4d_rf3_lda
            Xval_naoinfec_4d_rf3_lda=tempLDA_Xval_naoinfec_4d_rf3_lda
            Xval_infec_7d_rf3_lda=tempLDA_Xval_infec_7d_rf3_lda
            Xval_naoinfec_7d_rf3_lda=tempLDA_Xval_naoinfec_7d_rf3_lda   
            Xtreino_val_infec_4d_lda=tempLDA_Xtreino_val_infec_4d_lda
            Xtreino_val_naoinfec_4d_lda=tempLDA_Xtreino_val_naoinfec_4d_lda
            Xtreino_val_infec_7d_lda=tempLDA_Xtreino_val_infec_7d_lda
            Xtreino_val_naoinfec_7d_lda=tempLDA_Xtreino_val_naoinfec_7d_lda

        else:    
            Xtreino_todosd_lda=np.concatenate((Xtreino_todosd_lda,tempLDA_Xtreino_todosd_lda),axis=1)          
            Xval_infec_todosd_lda=np.concatenate((Xval_infec_todosd_lda,tempLDA_Xval_infec_todosd_lda),axis=1)
            Xval_naoinfec_todosd_lda=np.concatenate((Xval_naoinfec_todosd_lda,tempLDA_Xval_naoinfec_todosd_lda),axis=1) 

            Xval_infec_4d_rf3_lda=np.concatenate((Xval_infec_4d_rf3_lda,tempLDA_Xval_infec_4d_rf3_lda),axis=1)
            Xval_naoinfec_4d_rf3_lda=np.concatenate((Xval_naoinfec_4d_rf3_lda,tempLDA_Xval_naoinfec_4d_rf3_lda),axis=1)
            Xval_infec_7d_rf3_lda=np.concatenate((Xval_infec_7d_rf3_lda,tempLDA_Xval_infec_7d_rf3_lda),axis=1)
            Xval_naoinfec_7d_rf3_lda=np.concatenate((Xval_naoinfec_7d_rf3_lda,tempLDA_Xval_naoinfec_7d_rf3_lda),axis=1)   

            Xtreino_val_infec_4d_lda=np.concatenate((Xtreino_val_infec_4d_lda,tempLDA_Xtreino_val_infec_4d_lda),axis=1)
            Xtreino_val_naoinfec_4d_lda=np.concatenate((Xtreino_val_naoinfec_4d_lda,tempLDA_Xtreino_val_naoinfec_4d_lda),axis=1)
            Xtreino_val_infec_7d_lda=np.concatenate((Xtreino_val_infec_7d_lda,tempLDA_Xtreino_val_infec_7d_lda),axis=1)
            Xtreino_val_naoinfec_7d_lda=np.concatenate((Xtreino_val_naoinfec_7d_lda,tempLDA_Xtreino_val_naoinfec_7d_lda),axis=1)            


    LDAtodosdados_=LDA(n_components=1)
    Todosdados_Xtreino_todosd_lda= LDAtodosdados_.fit_transform(Xtreino_todosd, Ytreino_todosd)
    Xtreino_todosd_lda=np.concatenate((Xtreino_todosd_lda,Todosdados_Xtreino_todosd_lda),axis=1) 


    Todosdados_Xval_infec_todosd_lda=LDAtodosdados_.transform(Xval_infec_todosd)
    Xval_infec_todosd_lda=np.concatenate((Xval_infec_todosd_lda,Todosdados_Xval_infec_todosd_lda),axis=1)
    Todosdados_Xval_naoinfec_todosd_lda=LDAtodosdados_.transform(Xval_naoinfec_todosd)
    Xval_naoinfec_todosd_lda=np.concatenate((Xval_naoinfec_todosd_lda,Todosdados_Xval_naoinfec_todosd_lda),axis=1)

    Todosdados_Xval_infec_4d_rf3_lda=LDAtodosdados_.transform(Xval_infec_4d)
    Xval_infec_4d_rf3_lda=np.concatenate((Xval_infec_4d_rf3_lda,Todosdados_Xval_infec_4d_rf3_lda),axis=1)
    Todosdados_Xval_naoinfec_4d_rf3_lda=LDAtodosdados_.transform(Xval_naoinfec_4d)
    Xval_naoinfec_4d_rf3_lda=np.concatenate((Xval_naoinfec_4d_rf3_lda,Todosdados_Xval_naoinfec_4d_rf3_lda),axis=1)

    Todosdados_Xval_infec_7d_rf3_lda=LDAtodosdados_.transform(Xval_infec_7d)
    Xval_infec_7d_rf3_lda=np.concatenate((Xval_infec_7d_rf3_lda,Todosdados_Xval_infec_7d_rf3_lda),axis=1)
    Todosdados_Xval_naoinfec_7d_rf3_lda=LDAtodosdados_.transform(Xval_naoinfec_7d)
    Xval_naoinfec_7d_rf3_lda=np.concatenate((Xval_naoinfec_7d_rf3_lda,Todosdados_Xval_naoinfec_7d_rf3_lda),axis=1)

    Todos_Xtreino_val_infec_4d_lda=LDAtodosdados_.transform(Xtreino_val_infec_4d)
    Xtreino_val_infec_4d_lda=np.concatenate((Xtreino_val_infec_4d_lda,Todos_Xtreino_val_infec_4d_lda),axis=1)
    Todos_Xtreino_val_naoinfec_4d_lda=LDAtodosdados_.transform(Xtreino_val_naoinfec_4d)
    Xtreino_val_naoinfec_4d_lda=np.concatenate((Xtreino_val_naoinfec_4d_lda,Todos_Xtreino_val_naoinfec_4d_lda),axis=1)

    Todos_Xtreino_val_infec_7d_lda=LDAtodosdados_.transform(Xtreino_val_infec_7d)
    Xtreino_val_infec_7d_lda=np.concatenate((Xtreino_val_infec_7d_lda,Todos_Xtreino_val_infec_7d_lda),axis=1)
    Todos_Xtreino_val_naoinfec_7d_lda=LDAtodosdados_.transform(Xtreino_val_naoinfec_7d)
    Xtreino_val_naoinfec_7d_lda=np.concatenate((Xtreino_val_naoinfec_7d_lda,Todos_Xtreino_val_naoinfec_7d_lda),axis=1)


    classificador_todosd= svm.SVC(kernel='poly',gamma=tipo_gama)

    classificador_todosd.fit(Xtreino_todosd_lda, Ytreino_todosd)

    # Utilizando dados de 4 e 7 DPI #
    yhat=classificador_todosd.predict(Xval_infec_todosd_lda)
    yhat2=classificador_todosd.predict(Xval_naoinfec_todosd_lda)
    result1=accuracy_score(Yval_infec_todosd,yhat)
    result2=accuracy_score(Yval_naoinfec_todosd,yhat2)

    acuracia_infec_todosd.append(result1)
    acuracia_naoinfec_todosd.append(result2)

   # TPR e SPC apenas dos dados 4 DPI #
    yhat=classificador_todosd.predict(Xval_infec_4d_rf3_lda)
    yhat2=classificador_todosd.predict(Xval_naoinfec_4d_rf3_lda)
    result1=accuracy_score(Yval_infec_4d,yhat)
    result2=accuracy_score(Yval_naoinfec_4d,yhat2)
    infec_parcial_4d_todosd.append(result1)
    naoinfec_parcial_4d_todosd.append(result2)
 
    # TPR e SPC apenas dos dados 7 DPI #
    yhat=classificador_todosd.predict(Xval_infec_7d_rf3_lda)
    yhat2=classificador_todosd.predict(Xval_naoinfec_7d_rf3_lda)
    result1=accuracy_score(Yval_infec_7d,yhat)
    result2=accuracy_score(Yval_naoinfec_7d,yhat2)
    infec_parcial_7d_todosd.append(result1)
    naoinfec_parcial_7d_todosd.append(result2)



   # Dados de treino 4 DPI #
    yhat=classificador_todosd.predict(Xtreino_val_infec_4d_lda)
    yhat2=classificador_todosd.predict(Xtreino_val_naoinfec_4d_lda)
    result1=accuracy_score(Ytreino_val_infec_4d,yhat)
    result2=accuracy_score(Ytreino_val_naoinfec_4d,yhat2)
    acuracia_treino_infec_4d.append(result1)
    acuracia_treino_naoinfec_4d.append(result2)

   # Dados de treino 7 DPI #
    yhat=classificador_todosd.predict(Xtreino_val_infec_7d_lda)
    yhat2=classificador_todosd.predict(Xtreino_val_naoinfec_7d_lda)
    result1=accuracy_score(Ytreino_val_infec_7d,yhat)
    result2=accuracy_score(Ytreino_val_naoinfec_7d,yhat2)
    acuracia_treino_infec_7d.append(result1)
    acuracia_treino_naoinfec_7d.append(result2)

acuracia_infec_todosd=np.array(acuracia_infec_todosd)
acuracia_naoinfec_todosd=np.array(acuracia_naoinfec_todosd)
infec_parcial_4d_todosd=np.array(infec_parcial_4d_todosd)
naoinfec_parcial_4d_todosd=np.array(naoinfec_parcial_4d_todosd)
infec_parcial_7d_todosd=np.array(infec_parcial_7d_todosd)
naoinfec_parcial_7d_todosd=np.array(naoinfec_parcial_7d_todosd)
acuracia_treino_infec_4d=np.array(acuracia_treino_infec_4d)
acuracia_treino_naoinfec_4d=np.array(acuracia_treino_naoinfec_4d)
acuracia_treino_infec_7d=np.array(acuracia_treino_infec_7d)
acuracia_treino_naoinfec_7d=np.array(acuracia_treino_naoinfec_7d)

print('TPR (acerto infectados): %0.3f (desvio padrão : %0.3f)' % (acuracia_infec_todosd.mean(),acuracia_infec_todosd.std()))
print('SPR (acerto não infectados): %0.3f (desvio padrão : %0.3f)' % (acuracia_naoinfec_todosd.mean(),acuracia_naoinfec_todosd.std()))
print('TPR e SPC apenas dos dados 4 DPI: TPR: %0.3f (desvio padrão: %0.3f) || SPC : %0.3f (desvio padrão: %0.3f)' % (infec_parcial_4d_todosd.mean(),infec_parcial_4d_todosd.std(),naoinfec_parcial_4d_todosd.mean(),naoinfec_parcial_4d_todosd.std())) 
print('TPR e SPC apenas dos dados 7 DPI: TPR: %0.3f (desvio padrão: %0.3f) || SPC : %0.3f (desvio padrão: %0.3f)' % (infec_parcial_7d_todosd.mean(),infec_parcial_7d_todosd.std(),naoinfec_parcial_7d_todosd.mean(),naoinfec_parcial_7d_todosd.std()))

    
    
```
